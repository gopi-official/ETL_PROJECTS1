FEATURE SELECTION

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
import pandas as pd

# Ensure that x and y contain no negative values and no missing values
x = x.clip(lower=0)  # Set negative values to 0 (if there are any)
x = x.fillna(0)  # Replace missing values with 0 (if there are any)

# Ensure y is categorical (for classification)
y = y.astype('category')

# SelectKBest with chi2 as the score function
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(x, y)

# Get feature scores
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)

# Combine the scores with the feature names
featureScores = pd.concat([dfcolumns, dfscores], axis=1)
featureScores.columns = ['specs', 'scores']

# Show the top features
print(featureScores)
```
